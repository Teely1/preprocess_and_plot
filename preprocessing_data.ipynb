```python
# drop the coulmns, whos not nedded
drop_list=['Cabin','Ticket','PassengerId']

train_df = train_df.drop(drop_list,axis=1)
test_df = test_df.drop(drop_list,axis=1)
```

```python
# fill missing Embarked values in both datasets
train_df.Embarked.fillna('S',inplace=True)
test_df.Embarked.fillna('S',inplace=True)
# fill the missing values in the Age column
train_df.Age.fillna(28, inplace=True)
test_df.Age.fillna(28, inplace=True)
# fill the null Fare values in test dataset
train_df.Fare.fillna(test_df.Fare.median(), inplace=True)
test_df.Fare.fillna(test_df.Fare.median(), inplace=True)
```

```python
# combining train and test dataframes to work with them simultaneously
Combined_data = [train_df, test_df]
```

```python
# extracting various title in Names column
for dataset in Combined_data:
    dataset['Title'] = dataset.Name.str.extract(' ([A-Za-z]+)\.', expand=False)

# various titles extracted from the names
sns.countplot(y='Title',data=train_df)
```

```python
# merging some titles
for dataset in Combined_data:
    dataset['Title'] = dataset['Title'].replace(['Don', 'Rev', 'Lady','Dr', 'Major', 'Sir',
    'Col', 'Capt', 'Countess', 'Jonkheer'], 'Special')

    dataset['Title'] = dataset['Title'].replace({'Mlle':'Miss','Ms':'Miss','Mme':'Mrs'})
# overview of new titles
train_df.groupby('Title',as_index=False)['Survived'].mean().sort_values(by='Survived',
                                                                        ascending=False)
```

```python
# new titels plot
sns.countplot(y='Title',data=train_df)
```

```python
# new feature IsAlone from the SibSp and parch columns
for dataset in Combined_data:
    dataset["Family"] = dataset['SibSp'] + dataset['Parch']
    dataset["IsAlone"] = np.where(dataset["Family"] > 0, 0,1)
    dataset.drop('Family',axis=1,inplace=True)
train_df.head()
```

```python
# dropping the name,SibSP and parch
for dataset in Combined_data:
    dataset.drop(['SibSp','Parch','Name'],axis=1,inplace=True)
```

```python
# creating another feature if the passenger is a child
for dataset in Combined_data:
    dataset["IsChild"] = np.where(dataset["Age"] < 15, 1, 0)
```

```python
# converting categorical variables into numerical ones
train_df2 = pd.get_dummies(train_df,columns=['Pclass','Sex','Embarked', 'Title'],drop_first=True)
test_df2 = pd.get_dummies(test_df,columns=['Pclass','Sex','Embarked', 'Title'],drop_first=True)
train_df2.head()
```

```python
# creating age bands
train_df2['AgeBands'] = pd.qcut(train_df2.Age,4,labels=False)
test_df2['AgeBands'] = pd.qcut(test_df2.Age,4,labels=False)
train_df2.head()
```

```python
# creating fare bands
train_df2['FareBand'] = pd.qcut(train_df2.Fare,7,labels=False)
test_df2['FareBand'] = pd.qcut(test_df2.Fare,7,labels=False)
```

```python
# drop age and fare columns
train_df2.drop(['Age','Fare'],axis=1,inplace=True)
test_df2.drop(['Age','Fare'],axis=1,inplace=True)
```
